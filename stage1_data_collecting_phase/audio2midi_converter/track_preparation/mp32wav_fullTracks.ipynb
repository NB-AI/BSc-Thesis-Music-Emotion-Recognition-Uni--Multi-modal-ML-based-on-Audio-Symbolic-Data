```python
all_mp3_tracks = glob.glob('../../../stage0_provided_information/tracks/*.mp3')

for single_mp3 in all_mp3_tracks:
    print(single_mp3)
    name = single_mp3.split('.')[0]
    sound = AudioSegment.from_mp3(single_mp3)
    sound.export("full_wav_generated/" + name + ".wav", format="wav")
```
