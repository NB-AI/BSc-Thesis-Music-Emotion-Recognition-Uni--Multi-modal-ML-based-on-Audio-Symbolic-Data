```python
def rename_move_file(path:str):
    '''rename and remove files which where generated with spleeter'''

    
    file_name_split = path.split('/')
    storage_name = file_name_split[1] + '_' + file_name_split[-1] 
    
    # Rename file:
    laster = file_name_split[-1] 
    folder_to_save = file_name_split[0] +'/'+file_name_split[1] 
    new_name_path = folder_to_save + '/' + storage_name
    os.rename(path, new_name_path)
    
    # Move renamed file:
    file_name = file_name_split[-1]
    new_folder = file_name.split('.')[0]
    storage_place = new_folder + '/' + storage_name
    shutil.move(new_name_path, storage_place)
    return
    
  
```
