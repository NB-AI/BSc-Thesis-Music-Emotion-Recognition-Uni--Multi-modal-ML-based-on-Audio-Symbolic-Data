# 1. Split the wav files into instrumentally separated wav files

The spleeter module needs to be accessible for this purpose.

```python
all_wav_tracks = glob.glob('../track_preparation/accompaniment/*.wav')
failed_list = []

output = 'InstrumentallySplittedWav'
os.makedirs(output, exist_ok=True)

for single_wav in all_wav_tracks:
    print()
    print(single_wav)

    try:
        ! spleeter separate -p spleeter:5stems -o {output} {single_wav}
            
    except:
        name = (single_wav.split('/')[-1]).split('.')[0]
        failed_list.append(name)
```

# 2. convert separated voices into individual MIDI files

Restart notebook when you haven't activate an environement with all the requirements of the Wang audio2midi converter.

```python
output = 'InstrumentallySplittedWav'
all_splitted_wav = glob.glob(output + '/*/*.wav')

output_pred = 'InstrumentallySplittedMIDI/'

failed_list_pred = []
succeed_list_pred = []
for single_wav in all_splitted_wav:
    print()
    print(single_wav)
    
    mp3_code = single_wav.split('/')[-2]
    name = (single_wav.split('/')[-1]).split('.')[0]

    try:
        ! python audio2midi_Wang_module/audio2midi-main/scripts/inference_pop_style_transfer.py {single_wav} {output_pred+mp3_code} {name}
        succeed_list_pred.append(single_wav)
    except:
        failed_list_pred.append(name)
        
print(failed_list_pred)
```

## Giving second change by predicting again:

```python
aim_midi = set(['vocals.mid', 'piano.mid', 'drums.mid', 'bass.mid', 'other.mid'])

for single_path in glob.glob(output + '/*'):
    folder_midis = os.listdir(single_path)
    if len(folder_midis)<5:
        print()
        aim_midi_copy = aim_midi.copy()
        missing_midis = aim_midi_copy - set(folder_midis)
        for missing_midi in missing_midis:
            print()
            mp3_code = single_path.split('/')[-1]
            to_pred_path = output + '/' + mp3_code + '/' + missing_midi.split('.')[0] + '.wav'
            
            try:
                ! python audio2midi_Wang_module/audio2midi-main/scripts/inference_pop_style_transfer.py {to_pred_path} {single_path} {missing_midi.split('.')[0]}
            except:
                pass
```

# 3. Merge the midis together

```python
fusioned_path = 'GeneratedFusionedMIDI_Wang_afterInstrumentalSplit/'
os.makedirs(fusioned_path, exist_ok=True)

for single_folder in glob.glob(output_pred + '*'):
    merged_midi = MidiFile()
    folder_midis = os.listdir(single_folder)
    first_round = True
    for midi in folder_midis:
        to_open = single_folder + '/' + midi
        loaded_midi = MidiFile(f'{to_open}', clip=True)
        if first_round == True:
            # Copy time resolution. (If mid1 and mid2 have different
            # ticks_per_beat you will get timing problems.)
            merged_midi.ticks_per_beat = loaded_midi.ticks_per_beat
            first_round = False
        merged_midi.tracks += loaded_midi.tracks
    
    mp3_code = (single_folder.split('/'))[-1]
    merged_midi.save(f'{fusioned_path}{mp3_code}_merged.mid')
```
